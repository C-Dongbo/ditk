## Import DeepPath class

```python
import os
import sys

module_path = os.path.abspath(os.path.join('../../..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
from graph.completion.deepPath.deep_path import DeepPath

deeppath = DeepPath()
```

## Read_dataset
Must be run before any other method if not using the original papers dataset. If you are using the original papers dataset skip this step and set dataset_return to be {directory_name : list of relations to be trained on}

```python
input_file_path = "./data/sample_input.txt"
dataset_return = deeppath.read_dataset(input_file_path)
```

## Train model
Trains the model on a set of relations from the dataset_return object, optionally takes in a single relation in the options directory.

```python
deeppath.train(dataset_return)
```

```python
deeppath.train(dataset_return, options={"relation" : "concept_agentbelongstoorganization"})
```

## Predict results
Finds the paths to be used within the knowledge graph, optionally takes in a single relation in the options directory.

```python
prediction_info = deeppath.predict(dataset_return)
```

## Evaluate model
This evaluates the model, getting the Mean Average Precision. Optionally takes in the prediction_info, and must take in the dataset returned from the read_dataset method as benchmark_data. 

```python
deeppath.evaluate(benchmark_data = dataset_return)
```
